# CURATED REGIONAL EARTHQUAKE WAVEFORMS (CREW) DATASET
## Preliminary demo

```python
import h5py
import utils
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# open the data 
file = h5py.File('../Pn/100_CREW.hdf5','r') # each hdf5 file has a corresponding csv file with metadata
data = file['data']
example_names = list(data.keys())
```

## Plot 3 random examples

```python
indices = np.random.randint(low=0,high=len(example_names),size=(3,1))
for index in indices:
    fig = utils.plot_example(data[example_names[index[0]]])
```

### We can open the metadata file to filter the data, and then access said data

```python
df = pd.read_csv('100_CREW_metadata.csv')
```

### Find and plot an example which source is at a depth of more than 600 km

```python
tdf   = df[df['source_depth_km']>600]
tname = tdf['example_name'].to_list()[0]
fig = utils.plot_example(data[tname])
```

### Find and plot an example which source to receiver distance is between 10 and 11 degrees

```python
tdf   = df[(df['path_ep_distance_deg']>10)&(df['path_ep_distance_deg']<11)]
tname = tdf['example_name'].to_list()[0]
fig = utils.plot_example(data[tname])
```
